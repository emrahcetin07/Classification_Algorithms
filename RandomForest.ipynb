```python
%cd /gdrive
import os
os.chdir('/gdrive/MyDrive/Sınıflandırma')
!ls
import pandas as pd
import numpy as np
import matplotlib.pyplot  as plt
veriler = pd.read_csv('veri.csv')
#print(veriler)
from sklearn.model_selection import train_test_split#test size 0,33 yani %33 ünü test etcez  ve%67 eğitcez

#Boy, kilo ve  yaş  ile cinsiyet tahmini yapması isteniyor.
x=veriler.iloc[:,1:4].values #girdi olark verdiklerimiz(boy,kilo,yaş) bağımsız değişkenler
y=veriler.iloc[:,4:].values#çıktı olarak bu yani cinsiyet

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=0)
#x ve y sıralı olamlı girdi, çıktı diye
#print(x_train)#eğitim seti
#print("\n\n\n")
#print(x_test)
#print("\n\n\n")
#print(y_train)
#print("\n\n\n")
#print(y_test)
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(x_train)
X_test=sc.transform(x_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=10,criterion='entropy')
rfc.fit(X_train,y_train)
y_pred=rfc.predict(X_test)
cm=confusion_matrix(y_test,y_pred)
print(cm)
```
