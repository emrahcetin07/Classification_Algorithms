```python
import pandas as pd
import numpy as np
import matplotlib.pyplot  as plt
veriler = pd.read_csv('veri.csv')
print(veriler)
```

```python
from sklearn.model_selection import train_test_split#test size 0,33 yani %33 ünü test etcez  ve%67 eğitcez

#Boy, kilo ve  yaş  ile cinsiyet tahmini yapması isteniyor.
x=veriler.iloc[:,1:4].values #girdi olark verdiklerimiz(boy,kilo,yaş) bağımsız değişkenler
y=veriler.iloc[:,4:].values#çıktı olarak bu yani cinsiyet

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=0)
#x ve y sıralı olamlı girdi, çıktı diye
print(x_train)#eğitim seti
print("\n\n\n")
print(x_test)
print("\n\n\n")
print(y_train)
print("\n\n\n")
print(y_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(x_train)
X_test=sc.transform(x_test)
```

Logistic Reg. Model

```python
from sklearn.linear_model import LogisticRegression
logr=LogisticRegression(random_state=0)
logr.fit(X_train, y_train) #eğitim x den ye yi öğreniyor.
y_pred=logr.predict(X_test) #eğitime göre tahmin ediyor X test e sokuyor
print(y_pred)
print(y_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
```
